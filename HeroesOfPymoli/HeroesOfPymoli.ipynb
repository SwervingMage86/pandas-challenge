```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
HerosOfPymoli = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(HerosOfPymoli)
purchase_data
```

## Player Count

* Display the total number of players


```python
total_players = purchase_data['SN'].value_counts().count()
total_df = pd.DataFrame({'Total Players': [total_players]})
total_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
items = purchase_data["Item Name"].value_counts().count()
tot_rev = purchase_data["Price"].sum()
avg_price = tot_rev / purchase_data["Price"].count() 
tot_purch = purchase_data["Purchase ID"].count()

#items
#tot_rev
#avg_price
#tot_purch

sum_table1 = pd.DataFrame({"Number of Unique Items": [items],
                           "Total Revenue": [tot_rev],
                           "Total Purchases": tot_purch,
                           "Average Price": avg_price})

sum_table1["Total Revenue"] = sum_table1["Total Revenue"].map("${:.2f}".format)
sum_table1["Average Price"] = sum_table1["Average Price"].map("${:.2f}".format)

sum_table1

                        
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#purchase_data["Gender"].value_counts()

male = purchase_data.loc[purchase_data["Gender"] == 'Male', :]
male_tot = male["SN"].value_counts().count() 
male_per = male_tot / total_players

female = purchase_data.loc[purchase_data["Gender"] == 'Female', :]
female_tot = female["SN"].value_counts().count() 
female_per = female_tot / total_players

other = purchase_data.loc[purchase_data["Gender"] == 'Other / Non-Disclosed', :]
other_tot = other["SN"].value_counts().count() 
other_per = other_tot / total_players

gender = ['Male', 'Female', 'Other / Non-Disclosed']

sum_table2 = pd.DataFrame({'Gender': gender,
                           'Total Count' : [male_tot , female_tot , other_tot],
                           'Percentage of Players' : [male_per , female_per , other_per]})
sum_table2['Percentage of Players'] = sum_table2['Percentage of Players'].map('{:.2%}'.format)

sum_table2 = sum_table2.set_index('Gender', drop=True)
sum_table2
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
m_purch_count = male['Purchase ID'].count()
m_tot_val = male['Price'].sum()
m_avg_price = m_tot_val / m_purch_count
m_tot_person = m_tot_val / male_tot

f_purch_count = female['Purchase ID'].count()
f_tot_val = female['Price'].sum()
f_avg_price = f_tot_val / f_purch_count
f_tot_person = f_tot_val / female_tot

o_purch_count = other['Purchase ID'].count()
o_tot_val = other['Price'].sum()
o_avg_price = o_tot_val / o_purch_count
o_tot_person = o_tot_val / other_tot

sum_table3 = pd.DataFrame({'Gender' : gender,
                           'Purchase Count' : [m_purch_count, f_purch_count, o_purch_count],
                           'Average Purchase Price' : [m_avg_price, f_avg_price, o_avg_price],
                           'Total Purchase Value' : [m_tot_val, f_tot_val, o_tot_val],
                           'Average Total/Person' : [m_tot_person, f_tot_person, o_tot_person]})

sum_table3['Average Purchase Price'] = sum_table3['Average Purchase Price'].map('${:,.2f}'.format)
sum_table3['Total Purchase Value'] = sum_table3['Total Purchase Value'].map('${:,.2f}'.format)
sum_table3['Average Total/Person'] = sum_table3['Average Total/Person'].map('${:,.2f}'.format)
sum_table3 = sum_table3.set_index('Gender', drop=True)

sum_table3
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9,14,19,24,29,34,39,100]
bin_names = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

purchase_data['Age Group'] = pd.cut(purchase_data["Age"], bins, labels=bin_names)

purchase_data

```

```python
age_counts = purchase_data['Age Group'].value_counts()
age_per = age_counts / total_players

sum_table4 = pd.DataFrame({'Total Count' : age_counts,
                           'Percentage of Players' : age_per})

sum_table4['Percentage of Players'] = sum_table4['Percentage of Players'].map('{:.2%}'.format)
sum_table4 = sum_table4.reindex(bin_names)
sum_table4
```
